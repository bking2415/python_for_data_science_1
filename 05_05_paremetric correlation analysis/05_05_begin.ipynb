## Segment 5 - Starting with parametric methods in pandas and scipy

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams

import scipy
from scipy.stats.stats import pearsonr
```

```python
%matplotlib inline
# 8 inches wide and 4 inches tall
rcParams['figure.figsize'] = 8,4
plt.style.use('seaborn-whitegrid')
```

### The Pearson Correlation

```python
address = '/Users/brandonking/PycharmProjects/python_for_data_science_1/Data/mtcars.csv'

cars = pd.read_csv(address)

cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
```

```python
sb.pairplot(cars)
```

```python
X = cars[['mpg', 'hp', 'qsec', 'wt']]
sb.pairplot(X)
```

### Using scipy to calculate the Pearson correlation coefficient

```python
mpg = cars['mpg']
hp = cars['hp']
qsec = cars['qsec']
wt = cars['wt']

pearsonr_coefficient, p_value =pearsonr(mpg, hp)
print(f'PearsonR Correlation Coefficient {pearsonr_coefficient:.3f}')
```

```python
pearsonr_coefficient, p_value =pearsonr(mpg, qsec)
print(f'PearsonR Correlation Coefficient {pearsonr_coefficient:.3f}')
```

```python
pearsonr_coefficient, p_value =pearsonr(mpg, wt)
print(f'PearsonR Correlation Coefficient {pearsonr_coefficient:.3f}')
```

### Using pandas to calculate the Pearson correlation coefficient
