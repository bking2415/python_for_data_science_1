```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams

import scipy
from scipy.stats import spearmanr
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 14, 7
plt.style.use('seaborn-whitegrid')
```

### The Spearman Rank Correlation

```python
address = '/Users/brandonking/PycharmProjects/python_for_data_science_1/Data/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
```

```python
sb.pairplot(cars)
```

```python
X = cars[['cyl', 'vs', 'am', 'gear']]
sb.pairplot(X)
```

```python
cyl = cars['cyl']
vs = cars['vs']
am = cars['am']
gear = cars['gear']

spearmanr_coefficient, p_value = spearmanr(cyl, vs)

print(f'Spearman Rank Correlation Coefficient {spearmanr_coefficient:0.3f}')
```

```python
spearmanr_coefficient, p_value = spearmanr(cyl, am)

print(f'Spearman Rank Correlation Coefficient {spearmanr_coefficient:0.3f}')
```

```python
spearmanr_coefficient, p_value = spearmanr(cyl, gear)

print(f'Spearman Rank Correlation Coefficient {spearmanr_coefficient:0.3f}')
```

### Chi-square test for independence

```python
table = pd.crosstab(cyl, am)

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table.values)
print(f'Chi-square statistic {chi2:0.3f}, p_value {p:0.3f}')
```

```python
table = pd.crosstab(cyl, vs)

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table.values)
print(f'Chi-square statistic {chi2:0.3f}, p_value {p:0.3f}')
```

```python
table = pd.crosstab(cyl, gear)

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table.values)
print(f'Chi-square statistic {chi2:0.3f}, p_value {p:0.3f}')
```
