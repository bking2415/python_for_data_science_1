## Segment 4 - Creating labels and annotations

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
```

### Labeling plot features
#### The functional method

```python
x = range(1,10)
y = [1, 2, 3, 4, 0.5, 4, 3, 2, 1]
plt.bar(x,y)
# Creating labels for plots
plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
```

```python
z = [1, 2, 3, 4, 0.5]

veh_type = ['bicycle', 'motorbike', 'car', 'van', 'stroller']

plt.pie(z, labels=veh_type)
plt.show()
```

#### The object-oriented method

```python
address = '/Users/brandonking/PycharmProjects/python_for_data_science_1/Data/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
                    
mpg = cars.mpg

# Create a figure
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')
# Labels for x and y axes
ax.set_xlabel('car names')
ax.set_ylabel('mpg')
```

### Adding a legend to your plot
#### The functional method

```python
plt.pie(z)
plt.legend(veh_type, loc='best')
plt.show()
```

#### The object-oriented method

```python
# Create a figure
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')
# Labels for x and y axes
ax.set_xlabel('car names')
ax.set_ylabel('mpg')
# Create legend
ax.legend(loc='best')
                 
```

```python
# Create a figure
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')
# Labels for x and y axes
ax.set_xlabel('car names')
ax.set_ylabel('mpg')
# Create legend
ax.legend(loc='best')
# Set limits of y-axis
ax.set_ylim([0, 45])
# Annotate the data
ax.annotate('Toyota Corrola', xy=(19, 33.9), xytext=(21, 35),arrowprops=dict(facecolor='black', shrink=0.05))
```
